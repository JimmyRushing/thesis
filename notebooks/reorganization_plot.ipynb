```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [100,200,300,400,500,600,700,800]
y = [1.86,.58, .26, .064, .026, .01, .003, .0015]
error = [.0095,.003, .001, .0004, .0001, .00007, .00002, .000012]
fig = plt.figure(figsize = (8, 6), dpi=300)

ax = plt.axes()
ax.set_xlabel('$\mathregular{\lambda}$ - Reorganization energy (meV)',fontdict=dict(weight='bold'),fontsize= 16,labelpad =14)
ax.set_ylabel('Mobility ($\mathregular{cm^{2}}$/ Vs)',fontdict=dict(weight='bold'), fontsize= 16,labelpad =12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax.set_yscale('log')
#plt.ylim([-.5, 2])
#plt.scatter(x,y,marker ='x',s = 400,lw=5, color = 'dodgerblue')
p = ax.plot(x, y,'o', color = 'dodgerblue',linestyle='--', linewidth=2, markersize = 15);
ax2= fig.add_subplot(222)
ax2.set_xlabel("Marcus Hopping", fontsize = 12)
ax2.set_xticks([])
ax2.set_yticks([])

ax2.spines["left"].set_visible(False)
ax2.spines["bottom"].set_visible(False)

ax2.annotate("", xytext=(1.5, 0.5), xy=(1.515, .05),
            arrowprops=dict(arrowstyle="<->",lw=2))
ax2.annotate("", xytext=(2.5, 0.05), xy=(1.5, .05),
            arrowprops=dict(arrowstyle="-",linestyle = '--'))
ax2.annotate("$\mathregular{\lambda}$",(1.25,0.3), fontsize = 15)
a = np.linspace(0,4,100)
b = .5*(a-2.5)**2+.05
plt.plot(a,b,scaley = False, color = "darkorange", linewidth = 3)
c = np.linspace(0,4,100)
d = .5*(a-1.5)**2+.1
plt.plot(c,d,scaley = False,color = "darkorange", linewidth = 3)

#line = np.linspace(100,800,1000)
#ax.plot(line, line*0, color = "black")
```
