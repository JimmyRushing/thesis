```python
import matplotlib.pyplot as plt
```

```python
x = [100,150,200,250,300,350,400,450,500,550,600,650,700,750,800]
y = [ .0021,.003,.019,.08,.17,.23,.37,.45,.49,.42,.48,.55,.61,.6,.55]

fig = plt.figure(figsize = (5, 5))

ax = plt.axes()
ax.set_xlabel('Temperature(K)',fontdict=dict(weight='bold'),fontsize= 16,labelpad =14)
ax.set_ylabel('Mobility ($\mathregular{cm^{2}}$/ Vs)',fontdict=dict(weight='bold'), fontsize= 16,labelpad =12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.scatter(x,y,marker ='o',s = 100)
#p = ax.plot(x, y,'x', color = 'black',linestyle='--', linewidth=2, markersize = 15);
;
```

```python
fig.savefig('../src/figures/temp', bbox_inches = "tight", facecolor = "white")
```

```python
x = np.linspace(0,1, 1000)
y = np.log(1/x)

fig = plt.figure(figsize = (5, 5), dpi =300)
ax = plt.axes()
ax.set_xlabel('x',fontdict=dict(weight='bold'),fontsize= 16,labelpad =14)
ax.set_ylabel('ln(1/x)',fontdict=dict(weight='bold'), fontsize= 16,labelpad =12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(x,y, lw =4, color = "darkorange")
```

```python
fig.savefig('../src/figures/naturallog', bbox_inches = "tight", facecolor = "white")
```
